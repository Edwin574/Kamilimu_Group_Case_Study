Handle missing values

```python
# check for missing values
df.isnull().sum()
```

```python
# remove duplicates
df.duplicated().sum()
```

```python
# drop any rows with missing values
df.drop_duplicates(inplace=True)
```

```python
# Drop rows missing voltage/frequency (critical)
df = df.dropna(subset=['voltage', 'frequency'])
```

```python
# Fill missing values in categorical columns
df['power_source'] = df['power_source'].fillna('unknown')
df['power_source_est'] = df['power_source_est'].fillna('unknown')
```

```python
# Convert timestamp to datetime
df['time'] = pd.to_datetime(df['time'])
```

```python
# handle outliers in Voltage and Frequency
df = df[(df['voltage'] >= 100) & (df['voltage'] <= 300)]
df = df[(df['frequency'] >= 30) & (df['frequency'] <= 70)]
```

Convert data types

```python
df['time'] = pd.to_datetime(df['time'], errors='coerce')
df['voltage'] = pd.to_numeric(df['voltage'], errors='coerce')
df['frequency'] = pd.to_numeric(df['frequency'], errors='coerce')
```

Explore Hospital data


```python
print("Unique hospitals:", df['hospital'].nunique())
print("Top hospitals by data points:\n", df['hospital'].value_counts().head(10))
```

```python
# check for missing entries
print("Missing hospital entries:", df['hospital'].isnull().sum())
print("Blank hospital entries:", (df['hospital'].str.strip() == '').sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot hospital distribution
top_hospitals = df['hospital'].value_counts().nlargest(10).index
plt.figure(figsize=(10, 5))
sns.countplot(data=df[df['hospital'].isin(top_hospitals)], x='hospital', order=top_hospitals)
plt.title("Top 10 Hospitals by Data Points")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Cross-tab: power source usage by hospital
hospital_power = pd.crosstab(df['hospital'], df['power_source_est'])

# Optional: Normalize to see percentages
hospital_power_percent = hospital_power.div(hospital_power.sum(axis=1), axis=0)

# Plot heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(hospital_power_percent, cmap="YlGnBu", annot=True, fmt=".2f")
plt.title("Estimated Power Source Usage by Hospital (as %)")
plt.ylabel("Hospital")
plt.xlabel("Estimated Power Source")
plt.tight_layout()
plt.show()
```

***Dominant Power Sources***

Grid is the primary estimated power source across most hospitals, with usage percentages ranging from 0.64 (connaught) to 1.00 (wilberforce), indicating its widespread reliance as the main power supply.
Solar serves as a significant secondary source, particularly in hospitals with potentially unreliable grids, with notable usage in hastings (0.52), kambia (0.78), kabala (0.82), masanga (0.79), ola_during (0.79), and princess_christian (0.79).
Generator usage is minimal (0.00 to 0.22), suggesting it is rarely the primary power source.
Power Conditioner and Unknown categories show low or zero usage (0.00 to 0.33), indicating limited detection or classification challenges.

***Hospital-Specific Patterns***

wilberforce depends entirely on the grid (1.00), reflecting a highly reliable grid connection.
hastings exhibits a balanced mix (0.15 grid, 0.52 solar, 0.33 unknown), suggesting variability and potential classification issues.
kambia and kabala rely heavily on solar (0.78 and 0.82) with minimal grid (0.22 and 0.18), likely due to grid instability.
connaught shows dual reliance with significant grid (0.64) and some solar (0.28).
princess_christian displays a diverse mix (0.02 generator, 0.79 grid, 0.12 solar, 0.07 unknown), indicating varied power source usage.

***Variability in Usage***

Hospitals like bonthe, king_harman_road, magburaka, and makeni exhibit high grid usage (0.87 to 0.99) with minimal solar, suggesting stable grid dependency.
Hospitals with higher solar usage (e.g., kambia, kabala) are likely in regions with unreliable grid infrastructure, relying on solar as a primary or backup source.

***Potential Classification Issues***

The unknown category is prominent in hastings (0.33) and princess_christian (0.07), indicating difficulties in the rule-based algorithm, possibly due to overlapping voltage/frequency ranges or missing data.
Low generator and power_conditioner usage may reflect underutilization or misclassification, with generators potentially grouped into 'unknown' or 'grid'.
